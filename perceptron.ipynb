## Load the Iris Data Set

Extract only the input features $X$ and the labels $y$. Then clean the data as follows:
1. Collapse the classes `versicolor` and `virginica` as the negative class.
2. Make the class `setosa` the positive class.

```python
iris = load_iris()
X, y = iris["data"], iris["target"]
y[y == 1] = -1  # versicolor -> negative
y[y == 2] = -1  # virginica -> negative
y[y == 0] = 1   # setosa -> positive

# make a DataFrame for easy data manipulation and plotting
dat = pd.DataFrame(data=X,
                   columns=["Sepal length",
                            "Sepal width",
                            "Petal length",
                            "Petal width"])
dat["class"] = y
```

```python
scatter0 = sns.scatterplot(data=dat,
                           x="Sepal length", y="Petal length",
                           hue="class", style="class",
                           markers={1: "P", -1: "X"}, legend="full",
                           palette={1: "#006B84", -1: "#D66A00"})
scatter0.set_title("Iris Data", fontsize=20)

# scatter0.get_figure().savefig("imgs/perceptron_iris0.png")
scatter0
```

```python
scatter_with_sep = scatter0.plot([4, 8], [2.5, 2.0], "r",
                                 [4, 8], [2.0, 5.0], "r",
                                 [4, 8], [3.0, 1.0], "r")[0]\
                           .get_figure()
scatter_with_sep.savefig("imgs/perceptron_iris_linear_sep.png")
scatter_with_sep
```
